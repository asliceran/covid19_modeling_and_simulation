```python
import numpy as np
import matplotlib.pyplot as plt

# Global plot settings for clarity
plt.rcParams.update({'font.size': 12})
plt.rcParams["figure.figsize"] = [12, 7]

def SIR_Euler_Turkey(initial_conds, k, label_suffix=""):
    t0 = 0            # Day 0 = April 1, 2020
    t_end = 275       # End = December 31, 2020 (~275 days)
    h = 1             # Step size (1 day)

    steps = int((t_end - t0)/h + 1)
    t = np.linspace(t0, t_end, steps)
    S = np.zeros(steps)
    I = np.zeros(steps)
    R = np.zeros(steps)

    # Setting initial conditions for April 1, 2020
    S[0] = initial_conds[0]
    I[0] = initial_conds[1]
    R[0] = initial_conds[2]
    N = S[0] + I[0] + R[0]

    # Euler's Method Loop with Piecewise Logic
    for n in range(steps - 1):
        current_day = t[n]

        # --- PIECEWISE TRANSMISSION RATE (b) LOGIC ---
        # Period 1: Lockdown (April 1 - May 31)
        if current_day < 61:
            b = 0.11  # Intercity travel prohibited, curfews

        # Period 2: Normalization & Tourism (June 1 - Sept 30)
        elif 61 <= current_day < 183:
            b = 0.18  # "New Normal" begins, bans lifted, tourism increases

        # Period 3: Second Wave Restrictions (Oct 1 - Dec 31)
        else:
            b = 0.28  # Reintroduction of weekend curfews and dining limits

        # Differential equations
        dS = -b * S[n] * I[n] / N
        dI = (b * S[n] * I[n] / N) - (k * I[n])
        dR = k * I[n]

        # Euler Step Updates
        S[n+1] = S[n] + h * dS
        I[n+1] = I[n] + h * dI
        R[n+1] = R[n] + h * dR

    # Visualization
    plt.plot(t, S, label=f'Susceptible {label_suffix}', color='blue', alpha=0.7)
    plt.plot(t, I, label=f'Infected {label_suffix}', color='red', linewidth=2.5)
    plt.plot(t, R, label=f'Removed {label_suffix}', color='green', alpha=0.7)

    # Add vertical markers for policy shifts to help your explanation
    plt.axvline(x=61, color='gray', linestyle='--', label='June 1: Normalization')
    plt.axvline(x=183, color='gray', linestyle=':', label='Oct 1: Restrictions Return')

    plt.xlabel('Days since April 1, 2020')
    plt.ylabel('Population in Millions')
    plt.title('3-Phase SIR Model: COVID-19 in Turkey (2020)')
    plt.legend(loc='best', fontsize='small')
    plt.grid(True, alpha=0.2)

# --- RUNNING THE SIMULATION ---

# Initial Conditions (Turkey April 1, 2020)
# N = 83.6M, I = ~15.6k, R = ~0.6k
S0 = 83598073
I0 = 15679
R0 = 610
initial_vals = [S0, I0, R0]

# k = 1/14 (standard quarantine period)
recovery_rate = 0.071

# Call the function
SIR_Euler_Turkey(initial_conds=initial_vals, k=recovery_rate)
plt.show()
```
